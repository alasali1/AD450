```python
import requests
import lxml
import html5lib
import pandas as pd
from bs4 import BeautifulSoup

url = 'https://www.worldometers.info/coronavirus/country/us/#nav-yesterday'
response = requests.get(url)
data = response.text

soup = BeautifulSoup(data, 'html.parser')
table = soup.find(id="usa_table_countries_today")


df = pd.read_html(str(table), flavor='html5lib')[0]

remove = ["#", "Source", "Projections"]

df = df.loc[:, ~df.columns.isin(remove)]
dataframe = df.drop([df.index[0]])

dataframe.head(10)
```

```python
dataframe.rename(columns={'Tot\xa0Cases/1M pop': 'Tot Cases/1M pop'}, inplace=True)
dfcases= dataframe[["USAState", "Tot Cases/1M pop"]]
print(dfcases.head().sort_values(by=['Tot Cases/1M pop']))
print(dfcases['Tot Cases/1M pop'].describe())
```

```python
dfdeaths= dataframe[["USAState", "TotalDeaths"]]
print(dfdeaths.head().sort_values(by=['TotalDeaths']))
print(dfdeaths['TotalDeaths'].describe())
```

```python
dfdeathspop= dataframe[["USAState", "Deaths/1M pop"]]
print(dfdeathspop.head().sort_values(by=['Deaths/1M pop']))
print(dfdeathspop["Deaths/1M pop"].describe())
```
